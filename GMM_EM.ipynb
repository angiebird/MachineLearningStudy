```python
import numpy as np
import matplotlib.pyplot as plt
import time
import math
%matplotlib inline
```

### Utility function

```python
def Gaussian(x, mu, sigma):
    return 1./(sigma * (2*math.pi)**.5) * math.exp(-(x - mu)**2/(2*sigma**2))
```

```python
def Gaussian_plot(mu, sigma, ax, scale):
    x = np.arange(mu - 10 * sigma, mu + 10 * sigma, sigma / 10)
    y = [Gaussian(xi, mu, sigma) * scale for xi in x]
    ax.plot(x, y)
```

## Generate Data

```python
def gen_mixture_gaussian_data(gaussian_param_ls):
    s = None
    for mu, sigma, num in gaussian_param_ls:
        s_new = np.random.normal(mu, sigma, num)
        if s is not None:
            s = np.concatenate((s, s_new), axis = 0)
        else:
            s = s_new
    np.random.shuffle(s)
    return s
```

### EMM

#### E-step

```python
def update_probik_ls(xi_ls, muk_ls, sigmak_ls):
    N = xi_ls.shape[0]
    K = muk_ls.shape[0]
    probik_ls = np.zeros((N, K)) # probik_ls[i, k] = p(xi|theta_k)
    for i in range(N):
        xi = xi_ls[i]
        for k in range(K):
            probik_ls[i, k] = Gaussian(xi, muk_ls[k], sigmak_ls[k])
    return probik_ls
```

```python
def update_rik_ls(pik_ls, xi_ls, muk_ls, sigmak_ls):
    K = pik_ls.shape[0] # number of cluster
    N = xi_ls.shape[0] # number of data
    rik_ls = np.zeros((N, K)) 
    probik_ls = update_probik_ls(xi_ls, muk_ls, sigmak_ls)
    rik_ls = pik_ls * probik_ls
    sum_rik_ls = np.sum(rik_ls, axis = 1)
    rik_ls = (rik_ls.T / sum_rik_ls).T
    return rik_ls
```

#### M-step

update pik_ls, muk_ls, sigmak_ls

```python
def update_pik_ls(rik_ls):
    return np.mean(rik_ls, axis = 0)
```

```python
def update_muk_ls(rik_ls, xi_ls):
    rk_ls = np.sum(rik_ls, axis = 0).T
    muk_ls = (rik_ls.T.dot(xi_ls))/rk_ls
    return muk_ls
```

```python
def update_sigmak_ls(rik_ls, xi_ls, muk_ls):
    K = rik_ls.shape[1] # number of cluster
    sigmak_ls = np.zeros(K)
    rk_ls = np.sum(rik_ls, axis = 0).T
    for k in range(K):
        rikk = rik_ls[:, k]
        mukk = muk_ls[k]
        sigmak_ls[k] = rikk.dot(xi_ls * xi_ls)/np.sum(rikk) - mukk**2
    return sigmak_ls
    
```

#### Iterate E-M step - Maximum Likelihood

```python
def emm(xi_ls, K):
    pik_ls = np.ones(K) / K
    muk_ls = np.random.random(K)
    sigmak_ls = np.ones(K) / K
    iter_num = 8
    fig, ax_ls = plt.subplots(iter_num)
    ax = ax_ls[0]
    dummy = ax.hist(s, 30, normed= True)
    for k in range(K):
        Gaussian_plot(muk_ls[k], sigmak_ls[k]**.5, ax, pik_ls[k])
    result = []
    for it in range(1, iter_num):
        # E step
        rik_ls = update_rik_ls(pik_ls, xi_ls, muk_ls, sigmak_ls)

        # M step
        pik_ls = update_pik_ls(rik_ls)
        muk_ls = update_muk_ls(rik_ls, xi_ls)
        sigmak_ls = update_sigmak_ls(rik_ls, xi_ls, muk_ls)
        
        result.append((muk_ls, sigmak_ls, pik_ls, rik_ls))
        ax = ax_ls[it]
        dummy = ax.hist(s, 30, normed= True)
        for k in range(K):
            Gaussian_plot(muk_ls[k], sigmak_ls[k]**.5, ax, pik_ls[k])
    return result
        
```

#### Stable Example

```python
param_ls = []
param_ls.append((-2, .2, 400)) # mu1, sigma1, num1
param_ls.append((2, .2, 600)) # mu2, sigma2, num2
s = gen_mixture_gaussian_data(param_ls)
fig, ax = plt.subplots()
dummy = ax.hist(s, 30, normed= True)

for param in param_ls:
    mu = param[0]
    sigma = param[1]
    num = param[2]
    Gaussian_plot(mu, sigma, ax, num * 1. /len(s))
```

```python
result = emm(s, 2)
```
